```python
print('-'*30)
print('CADASTRE UMA PESSOA')
print('-'*30)
maior = homem = mulher = 0
while True:
  i = int(input('idade: '))
  s = ' '
  while s not in 'MF':
    s =str(input('Sexo: [F/M] ')).strip().upper()[0]
  print('-'*30)
  c = ' '
  while c not in 'SN':
    c = str(input('Quer continuar? [S/N] ')).strip().upper()[0]
  print('-='*15)
  if s == 'M':
    homem += 1
  elif i < 20:
    mulher += 1
  if i <= 18:
    maior += 1
  if c == 'N':
    break
print(f'{homem} são Homem \n{maior} são maiores de 18 anos \n{mulher} são mulheres com menos de 20 anos')
```
