```python
from random import randint
n = m = c = 0
while True:
  pi = str(input('Par ou Ímpar? [P/I]')).upper().strip()[0]
  n = int(input('Digite um Valor: '))
  m = randint(0,11)
  nm = n+m
  if pi == 'P':
    pi = 'PAR'
  if pi == 'I':
    pi = 'ÍMPAR'
  if nm % 2 == 0:
    PI = 'PAR'
  else:
    PI = 'ÍMPAR'
  print(f'Você jogou {n} e a máquina {m}. O total deu {nm}, esse número é {PI}.')
  if pi == PI:
      print('Você Ganhou!')
      c += 1
      print('-='*30)
  else:
    break
print (f'Par ou Ímpar encerrado ! Você venceu {c} vezes consecutivas.')
```

```python
from random import randint
v=0
while True:
  jogador = int(input( 'Diga Valor: '))
  computador = randint(0,11)
  total = jogador + computador
  tipo = ' '
  while tipo not in 'PI':
    tipo = str(input('Par ou Ímpar? [P/I]')).strip().upper()[0]
  print(f'Você jogou {jogador} e a máquina {computador}. Total de {total}.')
  print('DEU PAR' if total % 2 == 0 else 'DEU ÍMPAR')
  if tipo == 'P':
    if total % 2 == 0:
      print('Você VENCEU!')
      v += 1
    else:
      print('Você PERDEU!')
      break
  elif tipo == 'I':
    if total % 2 == 1:
      print('Você VENCEU!')
      v += 1
    else:
      print('Você PERDEU!')
      break
  print('Vamos Jogar novamente!')
  print('=-'*30)
print(f'GAME OVER ! Você venceu {v} vezes.')
```
