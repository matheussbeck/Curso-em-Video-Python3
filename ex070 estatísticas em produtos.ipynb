```python
print('-'*40)
print('LOJA SUPER BARATÃO')
print('-'*40)
total = tot1000 = 0
menor = ' '
min = 100000000000
while True:
  prod = str(input('Nome do Produto: ')).strip()
  preço = int(input('Preço: R$ '))
  total += preço
  if preço < min:
    min = preço
    menor = prod
  if preço >= 1000:
    tot1000 += 1
  c = ' '
  while c not in 'SN':
    c = str(input('Quer continuar ? [S/N]')).strip().upper()[0]
  if c == 'N':
    break
print(f'Você gastou R$ {total:.2f}')
print(f'{tot1000} produtos custam mais de R$ 1.000,00')
print(f'{menor} foi o produto mais barato e custou R$ {min:.2f}')
```
